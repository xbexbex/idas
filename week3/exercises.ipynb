# Exercise 2

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/hasy-data-labels.csv")
print(df)
```

```python
df = df.loc[(df['symbol_id'] >= 70) & (df['symbol_id'] <= 80)]
print(df)
```

```python
import matplotlib.pyplot as plt

imglist = []

for file in df["path"]:
    img = plt.imread("data/" + file)[:,:,0].flatten()
    imglist.append(img)
    
df["img"] = imglist
```

```python
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)

df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['img'].values, df['symbol_id'].values, test_size=0.2)

X_train = np.stack(X_train)
X_test = np.stack(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy = "most_frequent")

dummy.fit(X_train, y_train)
dummy.score(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

for i, n in enumerate(y_pred):
    if n != y_test[i]:
        plt.imshow(X_test[i].reshape(32,32))
        plt.show()

        print("Predicted: " + str(y_pred[i] - 70))
        print("Actual: " + str(y_test[i] - 70))
```

Especially 3s and 8s are troublesome. Both of the numbers are hard to write in a consistent manner, which causes trouble for the model that expects certain pixels to be in certain places. The other missclassified examples are also written in a manner that differs from the expected, e.g. by being mostly off center.
